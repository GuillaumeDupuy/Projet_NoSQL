```python
import pymongo
from pymongo import MongoClient
import json
from pprint import pprint
```

```python
client = MongoClient('localhost',27017)
db = client.ghibli
movies = db.film
people = db.people
locations = db.locations
vehicles = db.vehicles
species = db.species
print('connexion sucess')
```
